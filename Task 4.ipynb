```python
df=df.drop(0).reset_index(drop=True)
```

```python
## Renaming first four columns 
```

```python
df= df.rename(columns={df.columns[0] : "Year" ,
                 df.columns[1] : "Duration_Seconds" , df.columns[2] : "Age",df.columns[3] : "Gender"})
```

```python
df.head()
```

```python
## Filling missing values with None 
```

```python
fill_cols = [col for col in df.columns if 'Q7' in col or 'Q38' in col]
df[fill_cols] = df[fill_cols].fillna('None')
```

```python
cleaning_dict = {
    'Masterâ€™s degree': "Master's degree",
    'Bachelorâ€™s degree': "Bachelor's degree",
    'Bachelora€™s degree': "Bachelor's degree", 
    'Some college/university study without earning a bachelorâ€™s degree': "Some college/university study without earning a bachelor's degree"
}
df['Q4'] = df['Q4'].replace(cleaning_dict)
```

```python
df.isnull().mean() * 100
```

```python
## Drop other rows that have missing values
```

```python
df = df.dropna(subset=['Age', 'Q2', 'Q3', 'Q5'])
```

```python
## Converting duration_seconds to numeric 
```

```python
df['Duration_Seconds']=pd.to_numeric(df['Duration_Seconds'])
```

```python
df.dtypes
```

```python
## filling duration_seconds with the median of hall column 
```

```python
df['Duration_Seconds']=df['Duration_Seconds'].fillna(df['Duration_Seconds'].median())
```

```python
df.isnull().mean() * 100
```

```python
##  applying label encoding or mapping for categorical variables 
```

```python
edu_mapping = {
    'Bachelor’s degree': 1,
    'Master’s degree': 2,
    'Doctoral degree': 3,
    'Professional doctorate': 4,
    'I prefer not to answer': 0,
    'Some college/university study without earning a bachelor’s degree': 0
}
df['edu_score'] = df['Q4'].map(edu_mapping)
df['gender_code'] = df['Q2'].astype('category').cat.codes
```

```python
## Meaningfull Insights
```

```python
edu_behavior = df.Q4.value_counts().reset_index()
edu_behavior.columns = ["Education" , "Count"]
```

```python
edu_behavior
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=edu_behavior, x="Count", y="Education", palette="viridis")

plt.title('Respondent Education Distribution')
plt.show()
```

```python
sns.countplot(data=df, x="Age", palette="magma", 
              order=['18-21', '22-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-69', '70+'])

plt.title('Age Distribution of Respondents')
plt.xticks(rotation=45) 
plt.show()
```

```python
df['Gender'] = df['Gender'].replace({
    'Man': 'Male',
    'Woman': 'Female',
    'Non-binary, genderqueer, or gender non-conforming' : 'Nonbinary'
})
```

```python
Total_Participants_Gender = df['Gender'].value_counts().reset_index()
Total_Participants_Gender.columns=["Gender","Count"]
```
