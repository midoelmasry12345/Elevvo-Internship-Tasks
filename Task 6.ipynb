```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = "https://realpython.github.io/fake-jobs/"
response = requests.get(url)

print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
job_cards = soup.find_all("div", class_="card-content")

print(len(job_cards))
```

```python
job = job_cards[0]

title = job.find("h2", class_="title").text.strip()
company = job.find("h3", class_="company").text.strip()
location = job.find("p", class_="location").text.strip()
description = job.find("div", class_="content").text.strip()

print(title)
print(company)
print(location)
```

```python
jobs_data = []

for job in job_cards:
    title = job.find("h2", class_="title").text.strip()
    company = job.find("h3", class_="company").text.strip()
    location = job.find("p", class_="location").text.strip()
    description = job.find("div", class_="content").text.strip()
    
    jobs_data.append({
        "Job Title": title,
        "Company": company,
        "Location": location,
        "Description": description
    })
```

```python
df = pd.DataFrame(jobs_data)
df.head(100)
```

```python
def clean_text(text):
    text = re.sub(r"\s+", " ", text)
    return text.strip()

df["Job Title"] = df["Job Title"].apply(clean_text)
df["Company"] = df["Company"].apply(clean_text)
df["Location"] = df["Location"].apply(clean_text)
df["Description"] = df["Description"].apply(clean_text)
```

```python
skills = [
    "python", "sql", "excel", "power bi", "tableau",
    "machine learning", "data analysis", "pandas", "numpy"
]
```

```python
def extract_skills(description):
    description = description.lower()
    found_skills = [skill for skill in skills if skill in description]
    return ", ".join(found_skills)
```

```python
df["Skills"] = df["Description"].apply(extract_skills)
df.head(20)
```
